```python
#preview
jhop_daily_death.head()
```

```python
#Extracting Nigeria data from the John Hopkins data sets and putting it into a dataframe

nig_global = []
for index, rows in jhop_daily_confirmed.iterrows():
    if index == 'Nigeria':
        nig_global.append(rows)
        
for index, rows in jhop_daily_recovered.iterrows():
    if index == "Nigeria":
        nig_global.append(rows)
        
for index, rows in jhop_daily_death.iterrows():
    if index == "Nigeria":
        nig_global.append(rows)

#print(nig_global[0])

nig_global = pd.DataFrame(nig_global)
nig_global
```

```python
#  Notice that the Index is the same "Nigeria", Meanwhile, each row represents Confirmed cases, Recovered Cases 
# and Death cases respectively.

# Hence, we rename the Index as apropriate
ind = list(nig_global.index)
ind[0] = "Nigeria_Conf"
ind[1] = "Nigeria_rec"
ind[2] = "Nigeria_death"
nig_global.index = ind
```

```python
#preview
nig_global
```

```python
# We can proceed to drop the irelevant columns such as Province and the Lat and Long columns.
nig_global = nig_global.drop(["Province/State", "Lat", "Long"], axis = 1)
nig_global
```

```python
# For better presentation we can transopse the table to be more readable
nig_global = nig_global.transpose()
nig_global.tail()
```

#### Further look at the budget data
#### Create Tables to show change in Budget and Percent Change in the budget

```python
#Lets create a new column for the change in budget
budget_change = []
for index, row in budget_data.iterrows():
    diff = row[1] - row[2]
    budget_change.append(diff)
    

budget_data["Budget_change (Bn)"] = budget_change
budget_data.head()
```

```python
#Percentage budget change
percent_bc = []
for index, row in budget_data.iterrows():
    perc = (row[3]/row[1]) * 100
    perc2 = round(perc, 2)
    percent_bc.append(perc2)
    
budget_data["Percent_budget_change"] = percent_bc
budget_data.head()
```
